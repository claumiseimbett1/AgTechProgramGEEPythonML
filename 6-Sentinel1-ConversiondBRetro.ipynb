```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Cargar imagen de Sentinel-1 en formato GeoTIFF (valor en sigma0 lineal)
with rasterio.open('VH.tif') as src:
    sigma0 = src.read(1)
    profile = src.profile

# Evitar log de cero
sigma0[sigma0 <= 0] = np.nan

# Convertir a dB
sigma0_dB = 10 * np.log10(sigma0)

# Mostrar imagen
plt.imshow(sigma0_dB, cmap='gray')
plt.colorbar(label='Retrodispersión (dB)')
plt.title('Sentinel-1 VV en dB')
plt.show()
```
