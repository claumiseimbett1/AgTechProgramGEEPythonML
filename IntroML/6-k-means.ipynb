```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# ---------------------------
# 1. Generar datos de ejemplo
# ---------------------------
from sklearn.datasets import make_blobs

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# ---------------------------
# 2. Aplicar K-Means
# ---------------------------
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# ---------------------------
# 3. Métricas de calidad
# ---------------------------
inertia = kmeans.inertia_   # Suma de distancias cuadradas dentro del cluster
silhouette = silhouette_score(X, y_kmeans)  # Medida de separación entre clusters

print(f"Inercia (distancia interna total): {inertia:.2f}")
print(f"Coeficiente de Silhouette: {silhouette:.2f}")

# ---------------------------
# 4. Visualización
# ---------------------------
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Mostrar los centroides
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')

plt.title("K-Means Clustering")
plt.xlabel("Característica 1")
plt.ylabel("Caractisticas 2")
plt.show()
```
