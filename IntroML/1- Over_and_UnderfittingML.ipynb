```python
import matplotlib.pyplot as plt
import numpy as np

# Simulacin de errores
x = np.linspace(1, 10, 10)

# Underfitting: error alto en train y test, casi iguales
train_under = np.linspace(0.8, 0.75, 10)
test_under = np.linspace(0.82, 0.77, 10)

# Overfitting: train bajo, test alto
train_over = np.linspace(0.4, 0.1, 10)
test_over = np.linspace(0.6, 0.9, 10)

# Buen ajuste: train y test convergen
train_good = np.linspace(0.6, 0.2, 10)
test_good = np.linspace(0.65, 0.25, 10)

fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Underfitting
axs[0].plot(x, train_under, label="Error entrenamiento")
axs[0].plot(x, test_under, label="Error validacin")
axs[0].set_title("Underfitting (Subajuste)")
axs[0].set_xlabel("Complejidad del modelo")
axs[0].set_ylabel("Error")
axs[0].legend()
axs[0].grid(True)

# Overfitting
axs[1].plot(x, train_over, label="Error entrenamiento")
axs[1].plot(x, test_over, label="Error validacin")
axs[1].set_title("Overfitting (Sobreajuste)")
axs[1].set_xlabel("Complejidad del modelo")
axs[1].legend()
axs[1].grid(True)

# Buen modelo
axs[2].plot(x, train_good, label="Error entrenamiento")
axs[2].plot(x, test_good, label="Error validacin")
axs[2].set_title("Buen Ajuste (Generalizacin)")
axs[2].set_xlabel("Complejidad del modelo")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
